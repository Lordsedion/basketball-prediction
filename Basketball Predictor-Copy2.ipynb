```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import pandas as pd
import numpy as np
from selenium.webdriver.common.by import By
from time import sleep
```

```python
url = 'https://www.flashscore.com.ng/team/crvena-zvezda/6uAuZ6Pt/results/'
```

```python
driver = webdriver.Chrome(executable_path = r"C:\Users\user\Downloads\chromedriver_win32 (1)\chromedriver.exe")
driver.get(url)
ID = 'id'
```

```python
gamed = driver.find_elements(By.CLASS_NAME, 'ui-table__row')
teambase = []

for game in gamed:
    try:
        home_team = game.find_element(By.CLASS_NAME, 'tableCellParticipant__name').text
    except:
        continue
    
    event_item = {
        'home_team': home_team,
    }
    
    teambase.append(event_item)

dt = pd.DataFrame(teambase)
```

```python
games = driver.find_elements(By.CLASS_NAME, 'event__match--twoLine')
gamesbase = []

for game in games:
    try:
        home_team = game.find_element(By.CLASS_NAME, 'event__participant--home').text
        away_team = game.find_element(By.CLASS_NAME, 'event__participant--away').text    

        first_quarter_home = game.find_element("xpath", './/*[@class="event__part event__part--home event__part--1"]').text
        second_quarter_home = game.find_element("xpath", './/*[@class="event__part event__part--home event__part--2"]').text
        third_quarter_home = game.find_element("xpath", './/*[@class="event__part event__part--home event__part--3"]').text
        fourth_quarter_home = game.find_element("xpath", './/*[@class="event__part event__part--home event__part--4"]').text
        first_quarter_away = game.find_element("xpath", './/*[@class="event__part event__part--away event__part--1"]').text
        second_quarter_away = game.find_element("xpath", './/*[@class="event__part event__part--away event__part--2"]').text
        third_quarter_away = game.find_element("xpath", './/*[@class="event__part event__part--away event__part--3"]').text
        fourth_quarter_away = game.find_element("xpath", './/*[@class="event__part event__part--away event__part--4"]').text

        event_time = game.find_element("xpath", './/*[@class="event__time"]').text
    except:
        continue
    
    event_item = {
        'home_team': home_team,
        'away_team': away_team,
        'first_quarter_home': first_quarter_home,
        'second_quarter_home': second_quarter_home,
        'third_quarter_home': third_quarter_home,
        'fourth_quarter_home': fourth_quarter_home,
        'first_quarter_away': first_quarter_away,
        'second_quarter_away': second_quarter_away,
        'third_quarter_away': third_quarter_away,
        'fourth_quarter_away': fourth_quarter_away,
        'event_time': event_time,
    }
    
    gamesbase.append(event_item)

df = pd.DataFrame(gamesbase)
```
